# Pyber Analysis

* Overall Pyber operates their services in 3 different geography area: Urban, Suburban and Rural
* Urban city type has highest number of revenue (by fare), highest number of rides and highest number of drivers (employeed)
* In terms of comparing average fare: Rural areas have highest fare where as urban area has lowest fare.
* Number of drivers gradually increases from Rural going to Suburban going to Urban area 

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride_data_df = pd.merge(ride_data_df, city_data_df, on="city", how="inner")

# Display the data table for preview
city_ride_data_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
urban_city_data_df = city_data_df.loc[city_data_df['type'] == 'Urban']
urban_city_group_data_df = urban_city_data_df.groupby('city')
urban_city_group_data_df['driver_count'].sum().head()
```

```python
# Obtain the x and y coordinates for each of the three city types
urban_data_df = city_ride_data_df.loc[city_ride_data_df['type'] == 'Urban']
urban_city_data_df = city_data_df.loc[city_data_df['type'] == 'Urban']
suburban_data_df = city_ride_data_df.loc[city_ride_data_df['type'] == 'Suburban']
suburban_city_data_df = city_data_df.loc[city_data_df['type'] == 'Suburban']
rural_data_df = city_ride_data_df.loc[city_ride_data_df['type'] == 'Rural']
rural_city_data_df = city_data_df.loc[city_data_df['type'] == 'Rural']

# Group by each of the typr by city
urban_group_df = urban_data_df.groupby('city')
urban_city_group_data_df = urban_city_data_df.groupby('city')
suburban_group_df = suburban_data_df.groupby('city')
suburban_city_group_data_df = suburban_city_data_df.groupby('city')
rural_group_df = rural_data_df.groupby('city')
rural_city_group_data_df = rural_city_data_df.groupby('city')

# Urban data point
urban_rides = urban_group_df['ride_id'].count()
urban_average_fares = urban_group_df['fare'].mean()
urban_driver_count = urban_city_group_data_df['driver_count'].sum()*10

# Suburban data point
suburban_rides = suburban_group_df['ride_id'].count()
suburban_average_fares = suburban_group_df['fare'].mean()
suburban_driver_count = suburban_city_group_data_df['driver_count'].sum()*10

# Urban data point
rural_rides = rural_group_df['ride_id'].count()
rural_average_fares = rural_group_df['fare'].mean()
rural_driver_count = rural_city_group_data_df['driver_count'].sum()*10

# Build the scatter plots for each city types
plt.scatter(urban_rides, urban_average_fares, s=urban_driver_count, c='lightcoral', alpha=0.8, edgecolors='black', label='Urban')
plt.scatter(suburban_rides, suburban_average_fares, s=suburban_driver_count, c='lightskyblue', alpha=0.8, edgecolors='black', label='Suburban')
plt.scatter(rural_rides, rural_average_fares, s=rural_driver_count, c='gold', alpha=0.8, edgecolors='black', label='Rural')

# Incorporate the other graph properties
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.grid()

# Create a legend
lgnd = plt.legend(title="City Types")

#change the marker size manually for both lines
lgnd.legendHandles[0]._sizes = [40]
lgnd.legendHandles[1]._sizes = [40]
lgnd.legendHandles[2]._sizes = [40]

# Incorporate a text label regarding circle size
text_label = 'Note:\nCircle size correlates with the driver count per city.'
plt.gcf().text(0.92, 0.6, text_label)

# Save Figure
plt.savefig('Pyber-ride-sharing-data.png', bbox_inches='tight')
```

## Total Fares by City Type

```python
city_ride_group_df = city_ride_data_df.groupby('type')

# Calculate Type Percents
fares_by_type = city_ride_group_df['fare'].sum().sort_values(ascending=False)

# Build Pie Chart
city_type = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", "gold"]
explode = (0.1,0,0)

plt.pie(fares_by_type, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", shadow=True, startangle=280)
plt.title('% of Total Fares by City Type')

# Save Figure
plt.savefig('Pyber-total-fares-by-city-type.png', bbox_inches='tight')
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_by_type = city_ride_data_df['type'].value_counts()

# Build Pie Chart
plt.pie(rides_by_type, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", shadow=True, startangle=260)
plt.title('% of Total Rides by City Type')

# Save Figure
plt.savefig('Pyber-total-rides-by-city-type.png', bbox_inches='tight')
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
city_group_df = city_data_df.groupby('type')
drivers_by_type = city_group_df['driver_count'].sum().sort_values(ascending=False)

# Build Pie Chart
plt.pie(drivers_by_type, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", shadow=True, startangle=260)
plt.title('% of Total Drivers by City Type')

# Save Figure
plt.savefig('Pyber-total-drivers-by-city-type.png', bbox_inches='tight')
```
