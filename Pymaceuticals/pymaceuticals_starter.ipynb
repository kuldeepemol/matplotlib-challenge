# Pymaceuticals Analysis

* Pymaceuticals conduct their drug test on mouse to monitor Tumor growth for 45 days timepoint.
* They used 9 different drugs against one controlled placebo. 
* From the analysis below we were able to see 2 drugs(Ramicane and Capomulin) where the tumor size was decrease in the Mouse

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merge_df = pd.merge(mouse_drug_data_df, clinical_trial_data_df, on="Mouse ID", how="outer")

# Display the data table for preview
merge_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_group_df = merge_df.groupby(['Drug', 'Timepoint'])

# Convert to DataFrame
mean_tumor_volume_df = drug_group_df['Tumor Volume (mm3)'].mean()
mean_tumor_volume_df = mean_tumor_volume_df.reset_index()

# Preview DataFrame
mean_tumor_volume_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor_volume_df = drug_group_df['Tumor Volume (mm3)'].sem()
sem_tumor_volume_df = sem_tumor_volume_df.reset_index()
# Convert to DataFrame

# Preview DataFrame
sem_tumor_volume_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
drug_timepoint_mean_df = mean_tumor_volume_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
drug_timepoint_sem_df = sem_tumor_volume_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
drug_timepoint_mean_df.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(np.arange(0, len(drug_timepoint_mean_df)*5, 5),
             drug_timepoint_mean_df['Capomulin'],
             yerr=drug_timepoint_sem_df['Capomulin'],
             fmt="o-", color="r", label="Capomulin"
            )
plt.errorbar(np.arange(0, len(drug_timepoint_mean_df)*5, 5),
             drug_timepoint_mean_df['Infubinol'],
             yerr=drug_timepoint_sem_df['Infubinol'],
             fmt="^-", color="b", label="Infubinol"
            )
plt.errorbar(np.arange(0, len(drug_timepoint_mean_df)*5, 5),
             drug_timepoint_mean_df['Ketapril'],
             yerr=drug_timepoint_sem_df['Ketapril'],
             fmt="s-", color="g", label="Ketapril"
            )
plt.errorbar(np.arange(0, len(drug_timepoint_mean_df)*5, 5),
             drug_timepoint_mean_df['Placebo'],
             yerr=drug_timepoint_sem_df['Placebo'],
             fmt="D-", color="k", label="Placebo"
            )
plt.title('Tumor Response to Treatment')
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()
plt.legend(loc="best")

# Save the Figure
plt.savefig('Tumor-Response-to-Treament.png')
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_metastatic_df = drug_group_df['Metastatic Sites'].mean()

# Convert to DataFrame
mean_metastatic_df = mean_metastatic_df.reset_index()
#mean_metastatic_df = pd.DataFrame({'Metastatic Sites': mean_metastatic_df})

# Preview DataFrame
mean_metastatic_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_metastatic_df = drug_group_df['Metastatic Sites'].sem()

# Convert to DataFrame
sem_metastatic_df = sem_metastatic_df.reset_index()
#sem_metastatic_df = pd.DataFrame({'Metastatic Sites': sem_metastatic_df})

# Preview DataFrame
sem_metastatic_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_mean_df = mean_metastatic_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
metastatic_sem_df = sem_metastatic_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')

# Preview that Reformatting worked
metastatic_mean_df.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(np.arange(0, len(metastatic_mean_df)*5, 5),
             metastatic_mean_df['Capomulin'],
             yerr=metastatic_sem_df['Capomulin'],
             fmt="o-", color="r", label="Capomulin"
            )
plt.errorbar(np.arange(0, len(metastatic_mean_df)*5, 5),
             metastatic_mean_df['Infubinol'],
             yerr=metastatic_sem_df['Infubinol'],
             fmt="^-", color="b", label="Infubinol"
            )
plt.errorbar(np.arange(0, len(metastatic_mean_df)*5, 5),
             metastatic_mean_df['Ketapril'],
             yerr=metastatic_sem_df['Ketapril'],
             fmt="s-", color="g", label="Ketapril"
            )
plt.errorbar(np.arange(0, len(metastatic_mean_df)*5, 5),
             metastatic_mean_df['Placebo'],
             yerr=metastatic_sem_df['Placebo'],
             fmt="D-", color="k", label="Placebo"
            )
plt.title('Metastatic Spread During Treatment')
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.grid()
plt.legend(loc="best")

# Save the Figure
plt.savefig('Metastatic-spread-during-treatment.png')
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count_df = drug_group_df['Mouse ID'].count()

# Convert to DataFrame
mouse_count_df = mouse_count_df.reset_index()
mouse_count_df.rename(columns={'Mouse ID': 'Mouse Count'}, inplace=True)

# Preview DataFrame
mouse_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_pivot_df = mouse_count_df.pivot(index='Timepoint', columns='Drug', values='Mouse Count')

# Preview the Data Frame
mouse_count_pivot_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(np.arange(0, len(mouse_count_pivot_df)*5, 5),
         (mouse_count_pivot_df['Capomulin']/mouse_count_pivot_df['Capomulin'][0]) * 100,
         "ro-", label="Capomulin"
        )
plt.plot(np.arange(0, len(mouse_count_pivot_df)*5, 5),
         (mouse_count_pivot_df['Infubinol']/mouse_count_pivot_df['Infubinol'][0]) * 100,
         "b^-", label="Infubinol"
        )
plt.plot(np.arange(0, len(mouse_count_pivot_df)*5, 5),
         (mouse_count_pivot_df['Ketapril']/mouse_count_pivot_df['Ketapril'][0]) * 100,
         "gs-", label="Ketapril"
        )
plt.plot(np.arange(0, len(mouse_count_pivot_df)*5, 5),
         (mouse_count_pivot_df['Placebo']/mouse_count_pivot_df['Placebo'][0]) * 100,
         "kD-", label="Placebo"
        )
plt.title('Survival During Treatment')
plt.xlabel("Time (Days)")
plt.ylabel("Survial Rate (%)")
plt.grid()
plt.legend(loc="best")

# Save the Figure
plt.savefig('Survival-during-treatment.png')

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_percent_change = ((drug_timepoint_mean_df.iloc[-1] - drug_timepoint_mean_df.iloc[0])/drug_timepoint_mean_df.iloc[0]) * 100

# Display the data to confirm
tumor_percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple

# Splice the data between passing and failing drugs
pass_drugs = tumor_percent_change.loc[tumor_percent_change < 0]
fail_drugs = tumor_percent_change.loc[tumor_percent_change > 0]
tumor_percent_change.sort_values(inplace=True)

# Orient widths. Add labels, tick marks, etc. 
plt.bar(np.arange(0, len(pass_drugs)), pass_drugs, color="g")
plt.bar(np.arange(2, len(fail_drugs)+2), fail_drugs, color="r")
plt.title("Tumor Change over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.grid()
tick_locations = [value+0.5 for value in np.arange(0,10)]
plt.xticks(tick_locations, tumor_percent_change.index, rotation=90)
plt.yticks([-30, -20, 0, 20, 40, 60, 70])

# Use functions to label the percentages of changes
for i in range(0,10):
    raw_val = tumor_percent_change[i]
    val = "{:.1f}".format(tumor_percent_change[i])
    if (i<2):
        plt.text(x = i , y = raw_val-7, s = val, size = 12, ha='center', va='bottom')
    else:
        plt.text(x = i , y = raw_val+7, s = val, size = 12, ha='center', va='bottom')


# Save the Figure
plt.savefig('Tumor-change-over-45-day-treatment.png')
```
